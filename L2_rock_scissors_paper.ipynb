```python
import random
playcounter = 1
counter = 0

def user_input():
  user_choice = input("Enter your choice (rock, paper, scissors): ")
  computer_choices = ["rock", "paper", "scissors"]
  computer_choice = random.choice(computer_choices)
  print(f"\nYou chose {user_choice}, computer chose {computer_choice}.\n")
  value = game(user_choice, computer_choice)
  return value

def game(user_choice, computer_choice):
  if user_choice == computer_choice:
    print("It's a tie!\n\n")
    return "tie"

  elif user_choice == "rock":
    if computer_choice == "scissors":
        print("You win!\n\n")
        return "win"
    else:
        print("You lose.\n\n")
        return "lose"
  
  elif user_choice == "paper":
    if computer_choice == "rock":
        print("You win!\n\n")
        return "win"
    else:
        print("You lose.\n\n")
        return "lose"

  elif user_choice == "scissors":
    if computer_choice == "paper":
        print("You win!\n\n")
        return "win"
    else:
        print("You lose.\n\n")
        return "lose"

while playcounter < 4:
  print (f"\n===================ROUND {playcounter}===================\n")
  value = user_input()
  if value == "tie":
    playcounter -= 1
  elif value == "win":
    counter += 1
  elif value == "lose":
    counter += 0
  else:
    print ("Wrong input. Please select again.")
    playcounter -= 1

  playcounter += 1  

if counter >= 2:
  print(f"\n-->Since you win {counter} rounds, you win!")
elif counter == 1:
  print(f"\n-->Since you only win {counter} round, you lose.")
else:
  print(f"\n-->Since you did not win in all rounds, you lose.")
```
